```python
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_API_KEY")
print(" API Key loaded!")
```

```python
# These are our sample documents
# Think of these as pages in a book
# We will search through them using hybrid search

documents = [
    "LangChain is a framework for building AI applications using large language models.",
    "RAG stands for Retrieval Augmented Generation which finds documents and adds them to prompts.",
    "FAISS is a vector database made by Facebook for storing and searching embeddings.",
    "Pinecone is a cloud vector database that scales to billions of vectors in production.",
    "Weaviate supports hybrid search combining semantic and keyword search together.",
    "Qdrant is built with Rust making it the fastest vector database available.",
    "LangGraph controls the flow of multi agent workflows using nodes and edges.",
    "CrewAI orchestrates teams of specialized AI agents to complete complex tasks.",
    "Prompt Engineering is the art of writing better prompts to get better AI answers.",
    "Embeddings convert text into numbers so computers can understand meaning.",
]

print(" Documents created!")
print(f"Total documents: {len(documents)}")
print()
for i, doc in enumerate(documents):
    print(f"Doc {i+1}: {doc[:50]}...")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Create embeddings model
# This converts text to numbers (vectors)
# Runs locally — no API needed!
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print(" Embeddings model loaded!")
print("Model: all-MiniLM-L6-v2")
print("This converts text → numbers locally!")
```

```python
from langchain_community.vectorstores import FAISS

# Store all documents in FAISS
# FAISS converts each document to vector using embeddings
vectorstore = FAISS.from_texts(documents, embeddings)

print(" Documents stored in FAISS!")
print()

# Now search by MEANING
query = "Which database is fastest?"

print(f" Query: {query}")
print("-" * 40)

# similarity_search finds documents by meaning
results = vectorstore.similarity_search(query, k=3)

print("🔍 SEMANTIC SEARCH RESULTS:")
for i, doc in enumerate(results):
    print(f"\nResult {i+1}: {doc.page_content}")
```

```python
from rank_bm25 import BM25Okapi

# BM25 needs tokenized documents
# Tokenize = split each document into individual words
tokenized_docs = [doc.lower().split() for doc in documents]

# Create BM25 search engine
bm25 = BM25Okapi(tokenized_docs)

print(" BM25 Keyword Search ready!")
print()

# Search using keywords
query = "Which database is fastest?"
tokenized_query = query.lower().split()

print(f" Query: {query}")
print(f"Tokenized: {tokenized_query}")
print("-" * 40)

# Get BM25 scores for all documents
scores = bm25.get_scores(tokenized_query)

# Get top 3 results
import numpy as np
top_indices = np.argsort(scores)[::-1][:3]

print("🔍 KEYWORD SEARCH RESULTS (BM25):")
for i, idx in enumerate(top_indices):
    print(f"\nResult {i+1} (score: {scores[idx]:.3f}):")
    print(f"{documents[idx]}")
```

```python
# HYBRID SEARCH = Semantic + Keyword combined!
# We take results from both and merge them smartly

print(" HYBRID SEARCH")
print("="*50)

query = "Which database is fastest?"
print(f" Query: {query}")
print("-" * 40)

# Step 1 — Get Semantic Search results with scores
semantic_results = vectorstore.similarity_search_with_score(query, k=5)

print("Step 1 — Semantic Results:")
for doc, score in semantic_results:
    print(f"  Score: {score:.3f} | {doc.page_content[:50]}...")

# Step 2 — Get BM25 Keyword results
tokenized_query = query.lower().split()
bm25_scores = bm25.get_scores(tokenized_query)

print("\nStep 2 — BM25 Keyword Scores:")
for i, score in enumerate(bm25_scores):
    if score > 0:
        print(f"  Score: {score:.3f} | {documents[i][:50]}...")

# Step 3 — Combine scores (Hybrid!)
print("\nStep 3 — Combining scores...")

# Create dictionary to store combined scores
hybrid_scores = {}

# Add semantic scores
for doc, score in semantic_results:
    # Lower FAISS score = more similar (so we invert)
    hybrid_scores[doc.page_content] = 1 / (1 + score)

# Add BM25 scores
for i, score in enumerate(bm25_scores):
    doc_content = documents[i]
    if doc_content in hybrid_scores:
        # Document found in both — ADD scores together!
        hybrid_scores[doc_content] += score * 0.3
    else:
        hybrid_scores[doc_content] = score * 0.3

# Step 4 — Sort by combined score
sorted_results = sorted(
    hybrid_scores.items(),
    key=lambda x: x[1],
    reverse=True
)[:3]

print("\n🏆 HYBRID SEARCH FINAL RESULTS:")
for i, (doc, score) in enumerate(sorted_results):
    print(f"\nResult {i+1} (hybrid score: {score:.3f}):")
    print(f"{doc}")
```

```python
print(" RERANKING")
print("="*50)
print("Reranking = retrieve more → score again → keep best!")
print()

query = "Which database is fastest?"

# Step 1 — First retrieve MORE results than needed (10 instead of 3)
# This gives reranker more options to work with!
initial_results = vectorstore.similarity_search(query, k=10)

print(f"Step 1 — Retrieved {len(initial_results)} initial results")
print()

# Step 2 — Score each result manually
# How relevant is each document to the query?
def relevance_score(query, document):
    """
    Calculate how relevant a document is to the query
    Simple scoring: count matching important words
    """
    query_words = set(query.lower().split())
    doc_words = set(document.lower().split())

    # Remove common words that don't matter
    stop_words = {'is', 'the', 'a', 'an', 'which', 'for', 'and', 'or', 'in'}
    query_words = query_words - stop_words
    doc_words = doc_words - stop_words

    # Count matching words
    matches = query_words.intersection(doc_words)

    # Score = matches / total query words
    score = len(matches) / len(query_words) if query_words else 0
    return score

print("Step 2 — Scoring each result:")
scored_results = []
for doc in initial_results:
    score = relevance_score(query, doc.page_content)
    scored_results.append((doc.page_content, score))
    print(f"  Score: {score:.2f} | {doc.page_content[:50]}...")

# Step 3 — Rerank by score
reranked = sorted(scored_results, key=lambda x: x[1], reverse=True)

print("\n🏆 RERANKED FINAL RESULTS (Top 3):")
for i, (doc, score) in enumerate(reranked[:3]):
    print(f"\nResult {i+1} (rerank score: {score:.2f}):")
    print(doc)
```

```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams, PointStruct
import uuid

print(" QDRANT VECTOR DATABASE")
print("="*50)

# Step 1 — Create Qdrant client
# Running locally in memory — no server needed!
client = QdrantClient(":memory:")
print("Qdrant client created — running in memory!")

# Step 2 — Create a collection
# Collection = like a table in database
# vectors_config tells Qdrant about our vectors
client.create_collection(
    collection_name="ai_knowledge",  # name of our collection
    vectors_config=VectorParams(
        size=384,          # dimension of our vectors
        distance=Distance.COSINE  # how to measure similarity
    )
)
print(" Collection created: ai_knowledge")

# Step 3 — Convert documents to vectors
print("\nConverting documents to vectors...")
doc_vectors = embeddings.embed_documents(documents)
print(f" Created {len(doc_vectors)} vectors")
print(f"Each vector has {len(doc_vectors[0])} dimensions")
```

```python
# Step 4 — Store vectors in Qdrant
print("Storing vectors in Qdrant...")

# Create points — each point has:
# id = unique identifier
# vector = the numbers representing the document
# payload = original text we can retrieve later
points = []
for i, (doc, vector) in enumerate(zip(documents, doc_vectors)):
    points.append(
        PointStruct(
            id=i,              # unique id for each document
            vector=vector,     # 384 dimension vector
            payload={"text": doc}  # original text stored here
        )
    )

# Upload all points to Qdrant
client.upsert(
    collection_name="ai_knowledge",
    points=points
)

print(f"✅ Stored {len(points)} vectors in Qdrant!")
print()

# Check collection info
collection_info = client.get_collection("ai_knowledge")
print(f"Collection name: ai_knowledge")
print(f"Total vectors stored: {collection_info.points_count}")
```

```python
# Step 5 — Search in Qdrant!
print("🔍 SEARCHING IN QDRANT")
print("="*50)

query = "Which database is fastest?"
print(f"❓ Query: {query}")
print("-" * 40)

# Step 1 — Convert query to vector
query_vector = embeddings.embed_query(query)
print(f"✅ Query converted to vector ({len(query_vector)} dimensions)")

# Step 2 — Search Qdrant (new API uses query_points)
results = client.query_points(
    collection_name="ai_knowledge",
    query=query_vector,    # search using query vector
    limit=3                # return top 3 results
).points

# Step 3 — Show results
print("\n🏆 QDRANT SEARCH RESULTS:")
for i, result in enumerate(results):
    print(f"\nResult {i+1}:")
    print(f"Score: {result.score:.3f}")
    print(f"Text: {result.payload['text']}")
```
